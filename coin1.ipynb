```python
import pandas as pd
import csv
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv("1127_received.csv",usecols=['TransactionId','TransacionDate','ProdcutId','quantity'])
```

```python
df=df.dropna()
```

```python
df.columns=['Id','Date', 'ProductId', 'Quantity']
```

```python
df1=df
df1
```

```python
counts=df1['Id'].value_counts()
```

```python
df1=df1[df1['Id'].isin(counts[counts>=2].index)]
df1
```

```python
df1['start'],df1['time'] = zip(*df1['Date'].map(lambda x: x.split(' ')))
```

```python
date='2017-03-25'
end='2018-05-30'
```

```python



df1= df1[(df1['start']>=date)&(df1['start']<=end)]
```

```python
basket = (df1
          .groupby(['Id', 'ProductId'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Id'))
```

```python
def encode_units(x):
    	if x <= 0:
        	return 0
    	if x >= 1:
        	return 1
```

```python
basket_sets = basket.applymap(encode_units)

```

```python
frequent_itemsets =apriori(basket_sets, min_support=0.01, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, min_threshold=1)

rules['length'] = rules['antecedants'].apply(lambda x: len(x))
rules=rules[ (rules['length'] == 1)]
rules=rules.drop(['antecedent support','consequent support','support','length','conviction'],   axis=1)

```
